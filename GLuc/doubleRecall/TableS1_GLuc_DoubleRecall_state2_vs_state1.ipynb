```python
#@title 2. Extract *.match and *.QR from the Zip files
from zipfile import ZipFile
import gzip

with ZipFile(proteinAFileName, 'r') as zipAObj:
  zipAObj.extractall()

with ZipFile(proteinB1FileName, 'r') as zipB1Obj:
  zipB1Obj.extractall()



#cat all match files to all_match
AFolder = zipAObj.namelist()[0]
B1Folder = zipB1Obj.namelist()[0]

#print(AFolder, BFolder)
#Aall_match
print("The following *.match files are found:")
Aall_match = ''
AQR = ''
for a in zipAObj.namelist():
  if a.startswith('__MACOSX/'):
    continue
  if a.endswith('match.gz'):
    print(a)
    with gzip.open(a, 'rt') as matchObj:
      match_content = matchObj.read()
      Aall_match += match_content
  elif a.endswith('QR'):
    AQR = a

#Ball_match
B1all_match = ''
B1QR = ''
for b in zipB1Obj.namelist():
  if b.startswith('__MACOSX/'):
    continue
  if b.endswith('match.gz'):
    print(b)
    with gzip.open(b, 'rt') as matchObj:
      match_content = matchObj.read()
      B1all_match += match_content
  elif b.endswith('QR'):
    B1QR = b


print("\nThe following *.QR files are found:")
print(AQR)
print(B1QR)
```

```python
#@title 3. Double Recall Analysis
import re, sys

#@markdown Number of Identical Chains in the Sequence: (e.g. homodimer's N_chain = 2)
N_chain  = 1 #@param {type:"number"}
def getDiff(l1, l2_sub1, l2_sub2):
    diffList=[]
    for line1 in l1:
        #print(line1)
        if not line1.strip():
            continue

        line12 = line1.strip()
        #print("reading+" + line12.strip() + "+")
        line12_sub = re.sub(r'[\s+|\(|\)]', "", line12)

        inf21 = re.findall(line12_sub, l2_sub1)
        inf22 = re.findall(line12_sub, l2_sub2)
        #print("inf2" + ' '.join(inf21) + ' '.join(inf22))
        if ((not inf21) or (not inf22)): #not in recall vio list of both comformations
            diffList.append(line1.strip())
            #print("not in both, append")
    return diffList

def scanB(list, setOld, all_f3):
    hx1=[]
    h1h2=[]
    setB1 = set()

    for line in list: # form A
        if  line in setOld: # already check for b1
            continue
        if not line.strip():
            continue

        line2 = line.strip()
        line2_sub = re.sub(r'[\s+|\(|\)|\-|\>]', "", line2)
        #print("debug: reading" + line2_sub)

        uniqueHX1Match = 0
        for f3line in all_f3:
            if not f3line.strip():
                continue

            line3 = f3line.strip()
            line3_sub = re.sub(r'[\s+|\(|\)]', "", line3)
            #print(line3_sub+"\n"+line2_sub)
            inf2 = re.findall(line2_sub, line3_sub)
            if (inf2):

                #print("debug: line2 " + line2_sub)
                #print("debug: line3 " + line3_sub)
                #if re.search(r'poss 2*', f3line): #dimer
                #my_regex = r'poss ' + re.escape(N_chain_str) + r"!\d"
                if re.search(rf'poss {N_chain} \*', f3line):
                  uniqueHX1Match = 1

                hx11 = ''
                matchlist = f3line.split("\n")
                for hline in matchlist:
                   #print(hline)

                    atomH1re = re.compile(r'^HX1: atom\:\s+\d+\s+(\d+[A-Z]\-[A-Z]+\d*)\s+')
                    atom1 = atomH1re.findall(hline)

                    #print(atom1)
                    if (atom1):
                        #hx1.append(atom1[0])
                        hx11 = atom1[0]
                    else:
                        #mid distance < 5 ang
                        disre = re.compile(r'(\d+\.*\d*)\(mid\)')
                        dis = disre.findall(hline)

                        #print(dis)
                        #if (dis):
                        #    print(float(dis[0]))
                        if (dis and float(dis[0]) <= 5.000001): #explained
                            #print("debug: f3line " + f3line)
                            atomH2re = re.compile(r'^   atom\:\s+\d+\s+(\d+[A-Z]\-[A-Z]+\d*)\s+')
                            atom2 = atomH2re.findall(hline)

                            if (hx11 and atom2[0]):
                                h1h2.append([hx11, atom2[0]])
                                print(uniqueHX1Match, hx11, atom2[0], line2) #output
                                setB1.add(line2)
                                if (uniqueHX1Match):
                                  #print("f3line " + f3line)
                                  hx1.append(hx11)
                            #list.insert(hx1, atom)
                break
    return(hx1, h1h2, setB1)

def getH1H2(list, all_f31, all_f32):
    hx1B1=[]
    h1h2B1=[]
    setB1 = set()
    setB2 = set()
    (hx1B1, h1h2B1, setB2) = scanB(list, setB1, all_f31)

    hx1B2=[]
    h1h2B2=[]
    setB3 = set()
    (hx1B2, h1h2B2, setB3) = scanB(list, setB2, all_f32)

    return (hx1B1+hx1B2, h1h2B1+h1h2B2)

with open(AQR, 'r') as f1:
  all_AQR = f1.read()

with open(B1QR, 'r') as f2:
  all_B1QR = f2.read()

all_AQR_sub = re.sub(r'[\s+|\(|\)]', "", all_AQR)
all_B1QR_sub = re.sub(r'[\s+|\(|\)]', "", all_B1QR)

lstA = all_AQR.splitlines()
lstB1 = all_B1QR.splitlines()

lstA_notlstB = getDiff(lstA, all_B1QR_sub, "")

lstB1_notlstA = getDiff(lstB1, all_AQR_sub, "")

# this is good
lstB_notlstA = lstB1_notlstA # in both notB1 and notB2
#print(lstB_notlstA)

print("Protein A Recall Violations, but <5 in Protein B")
(lstAHX1, lstAH1H2) = getH1H2(lstA_notlstB, B1all_match.split("#\n#\n"),  [])
print(lstAHX1, lstAH1H2)
print("Protein B Recall Violations, but <5 in Protein A")
(lstBHX1, lstBH1H2) = getH1H2(lstB_notlstA, Aall_match.split("#\n#\n"), [])
print(lstBHX1, lstBH1H2)



```

```python
#@title 4. Make Plots

#@markdown ####4.1. Definition<br>
#@markdown H1 dimension - the H dimension with attached C or N<br>
#@markdown H2 dimension - the indirection H dimension<br>
#@markdown H1 Match - the X-axis for residue with matches on H1 dimension<br>
#@markdown H2 Match - the Y-axis for residue with matches on H2 dimension<br>
#@markdown Unique H1 - the number of H1s that have only one H matched from the chemical shift list<br>

#@markdown ####4.2. Plots<br>
#@markdown Top plot - a bar chart counting number of unique H1 NOESY peaks that can be exclusively explained by one model and not the other<br>
#@markdown Bottom plot - a contact map where a NOESY peak based on H-H distance from the RPF zip file for residue i to j exclusively explained by one model and not the other<br>

#@markdown ####4.3. Settings
import plotly.graph_objects as go

Length = 170 #@param {type:"number"}
#@markdown - the length for the "H1/H2 Match" axes
MaxH1 = 50 #@param {type: "number"}
#@markdown - the length for "unique H1" axis
Model_A_Name = 'class 2' #@param {type:"string"}
Model_A_Color = 'blue' #@param {type: "string"}

Model_B_Name = 'class1' #@param {type:"string"}
Model_B_Color = 'orange' #@param {type:"string"}


#H1plot
lstAH1Index = []
for h1 in lstAHX1:
  rindex = re.findall(r'^\d+', h1)
  lstAH1Index.append(int(rindex[0]))

lstBH1Index = []
for h1 in lstBHX1:
  rindex = re.findall(r'^\d+', h1)
  lstBH1Index.append(int(rindex[0]))

#print(lstBH1Index, lstAH1Index)
lstsetB = set(lstBH1Index)
lstsetA = set(lstAH1Index)

#print(lstsetA, lstsetB)

#2D plot
Bplot = []
h1Index = []

Blongrange = 0
BlongrangeHX1 = 0
Alongrange = 0
AlongrangeHX1 = 0

for a in lstAH1H2:
  h1Index = re.findall(r'^\d+', a[0])
  h2Index = re.findall(r'^\d+', a[1])
  #print(h1Index[0], h2Index[0])
  Bplot.append((int(h1Index[0]), int(h2Index[0])))
  if (abs(int(h1Index[0])-int(h2Index[0])) >= 5):
    Blongrange +=1
    #print(h1Index[0], lstsetA)
    if int(h1Index[0]) in lstsetA:
      BlongrangeHX1 +=1
  #AH1[h1Index[0]] = AH1[h1Index[0]]+1

Aplot = []
h2Index = []
for b in lstBH1H2:
  h1Index = re.findall(r'^\d+', b[0])
  h2Index = re.findall(r'^\d+', b[1])
  #print(h1Index[0], h2Index[0])
  Aplot.append((int(h1Index[0]), int(h2Index[0])))
  if (abs(int(h1Index[0])-int(h2Index[0])) >= 5):
    Alongrange +=1
    if int(h1Index[0]) in lstsetB:
      AlongrangeHX1 +=1
  #AH1[h1Index[0]] = AH1[h1Index[0]]+1

#print(lstBH1Index)
#stat

print("Number of H-Hpairs explained by ", Model_B_Name, ":", len(lstAH1H2), "longrange:", Blongrange, "(unique H1:", BlongrangeHX1,")")
print("Number of H-Hpairs explained by ", Model_A_Name, ":", len(lstBH1H2), "longrange:", Alongrange, "(unique H1:", AlongrangeHX1,")")

x1 = [10, 18, None, 23, 28, None, 36, 51, None, 53, 62, None, 66, 73, None,
     102, 107, None, 108, 120, None, 124, 133, None, 137, 144, None, 146, 154, None,159,162]
y1 = [170,170, None, 170,170, None, 170,170, None, 170,170, None, 170,170, None,
     170,170, None, 170,170, None, 170,170, None, 170,170, None, 170,170, None, 170,170]


traceH1 = go.Scatter(
    x = x1,
    y = y1,

    marker = dict(
        #size = 25,
        color="gray"
    )
)

traceH2 = go.Scatter(
    x = y1,
    y = x1,

    marker = dict(
        #size = 25,
        color="gray"
    )
)

trace1 = go.Scatter(
      x=list(zip(*Aplot))[0],
      y=list(zip(*Aplot))[1],
      name = Model_A_Name,
      #opacity = 0.5,
      mode="markers",
      marker = dict(
          size = 5,
          color=Model_A_Color
      )
)
trace2 = go.Scatter(
    x=list(zip(*Bplot))[0],
    y=list(zip(*Bplot))[1],
    name = Model_B_Name,
    #opacity = 0.5,
    mode="markers",
    marker = dict(
       size = 5,
       color=Model_B_Color
    )
)

trace3 = go.Histogram(
    x=lstBH1Index,
    yaxis="y2",
    xbins={"size": 1},
    name = Model_A_Name,
    showlegend=False,
    marker = dict(
      #size = 25,
      color=Model_A_Color
    )
 )
trace4 = go.Histogram(
    x=lstAH1Index,
    yaxis="y2",
    xbins={"size": 1},
    name = Model_B_Name,
    showlegend=False,
    marker = dict(
        #size = 25,
        color=Model_B_Color
    )
)

data= [traceH1, traceH2, trace1, trace2, trace3, trace4]

layout = go.Layout(
    margin=dict(l=20, r=20, t=20, b=20),
    autosize=False,
    height=600,
    xaxis = dict(
      range = [0,Length+5],
      autorange=False,
      title = 'H1 Match',
      showgrid = True,
      zeroline = True,
      showline = True,
      dtick = 10,
      tick0 = 0,
     ),
    yaxis = dict(
      range = [0,Length+5],
      autorange=False,
      domain = [0,0.85],
      title = 'H2 Match',
      showgrid = True,
      zeroline = True,
      showline = True,
      dtick = 10,
      tick0 = 0,
    ),
    yaxis2 = dict(
      range = [0,MaxH1],
      domain=[0.85,1],
      title = 'Unique H1',
      zeroline = True,
    )
    #paper_bgcolor="LightSteelBlue"
)

fig = go.Figure(data=data, layout=layout)

fig.add_hrect(y0=146, y1=164, line_width=0, fillcolor="red", opacity=0.2)
fig.add_vrect(x0=146, x1=164, line_width=0, fillcolor="red", opacity=0.2)

fig.add_hrect(y0=7, y1=19, line_width=0, fillcolor="blue", opacity=0.2)
fig.add_vrect(x0=7, x1=19, line_width=0, fillcolor="blue", opacity=0.2)

fig.add_hrect(y0=79, y1=101, line_width=0, fillcolor="green", opacity=0.2)
fig.add_vrect(x0=79, x1=101, line_width=0, fillcolor="green", opacity=0.2)


fig.show()

```
